```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

def scrape_cnbc_headlines(url="https://www.cnbc.com/world/?region=world"):
    """
    Scrapes CNBC to extract financial news headlines.
    Returns a list of headline strings.
    """
    response = requests.get(url)
    if response.status_code != 200:
        print(f"Failed to retrieve webpage. Status code: {response.status_code}")
        return []

    soup = BeautifulSoup(response.text, "html.parser")
    headlines = []

    # Adjust selector if CNBC changes layout
    for item in soup.select("a.LatestNews-headline"):
        headline = item.get_text(strip=True)
        if headline:
            headlines.append(headline)

    return headlines

def analyze_headlines(headlines):
    """
    Uses VaderSentiment to analyze sentiment of each headline.
    Returns a DataFrame with sentiment scores.
    """
    analyzer = SentimentIntensityAnalyzer()
    results = []
    for hl in headlines:
        scores = analyzer.polarity_scores(hl)
        results.append(scores)
    return pd.DataFrame(results)

# 1) Scrape the headlines
headlines = scrape_cnbc_headlines()
print(f"Scraped {len(headlines)} headlines.")
if len(headlines) > 0:
    # Show first 5
    print("Sample headlines:", headlines[:5])

# 2) Analyze sentiment
if len(headlines) > 0:
    sentiment_df = analyze_headlines(headlines)
    # Combine headlines + sentiment
    df = pd.DataFrame({"Headline": headlines})
    df = pd.concat([df, sentiment_df], axis=1)

    print("\n=== Sentiment Analysis Results ===")
    print(df.head())

    # Save to CSV in Colab (optional)
    df.to_csv("sentiment_results.csv", index=False)
    print("\nSaved results to 'sentiment_results.csv'!")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_cnbc_headlines(url="https://www.cnbc.com/world/?region=world"):
    # Add a "User-Agent" header to mimic a regular web browser
    headers = {
        "User-Agent": (
            "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
            "AppleWebKit/537.36 (KHTML, like Gecko) "
            "Chrome/91.0.4472.124 Safari/537.36"
        )
    }

    response = requests.get(url, headers=headers)
    if response.status_code != 200:
        print(f"Failed to retrieve webpage. Status code: {response.status_code}")
        return []

    soup = BeautifulSoup(response.text, "html.parser")
    headlines = []

    # Adjust the selector if CNBC changes its layout
    for item in soup.select("a.LatestNews-headline"):
        headline = item.get_text(strip=True)
        if headline:
            headlines.append(headline)

    return headlines
```

```python
headlines = scrape_cnbc_headlines()
print(f"Scraped {len(headlines)} headlines.")
# Print the first few
print(headlines[:5])
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd

# Assume your headlines are stored in a variable named `headlines`
analyzer = SentimentIntensityAnalyzer()

# We'll store sentiment scores for each headline in a list
scores = []
for hl in headlines:
    sentiment = analyzer.polarity_scores(hl)
    scores.append(sentiment)

# Combine headlines & sentiment scores into a DataFrame
df = pd.DataFrame({'Headline': headlines})
sentiment_df = pd.DataFrame(scores)
result_df = pd.concat([df, sentiment_df], axis=1)

# Display first few rows
print(result_df.head())

# (Optional) Save to CSV
result_df.to_csv("sentiment_results.csv", index=False)
print("Saved sentiment_results.csv!")
```

```python
import pandas as pd
from google.colab import files

# Assuming you already have a DataFrame called `result_df` with columns:
# ["Headline", "neg", "neu", "pos", "compound"]

# 1. Rename columns for clarity
result_df.rename(columns={
    'neg': 'Negative_Score',
    'neu': 'Neutral_Score',
    'pos': 'Positive_Score',
    'compound': 'Overall_Sentiment'
}, inplace=True)

# 2. Add a human-readable label (Positive, Negative, or Neutral)
def label_sentiment(compound_score):
    if compound_score >= 0.05:
        return "Positive"
    elif compound_score <= -0.05:
        return "Negative"
    else:
        return "Neutral"

result_df["Sentiment_Label"] = result_df["Overall_Sentiment"].apply(label_sentiment)

# 3. Save the updated DataFrame to a CSV file with a clear name
csv_filename = "final_sentiment_results.csv"
result_df.to_csv(csv_filename, index=False)
print(f"Saved renamed and labeled results to '{csv_filename}'!")

# 4. Download the CSV to your local machine
files.download(csv_filename)
```
